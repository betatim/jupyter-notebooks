```python
#import libraries required for analysis
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import  DateFormatter
%matplotlib inline
```

```python
projectid = "ebmdatalab"
#obtain data for lowest 10 CCG prescribers of trimethoprim
lowest_10_percent_df = pd.read_gbq("""
SELECT
  month,
  IEEE_DIVIDE(SUM(
      CASE
        WHEN SUBSTR(lowest.bnf_code,1,9)='0501080W0' THEN items
        ELSE 0 END), SUM(items)) AS lowest_10_trimeth_percent
FROM
  `ebmdatalab.richard.trimethoprim_data_lowest_10` AS lowest
WHERE
  month>='2014-01-01'
GROUP BY
  month
ORDER BY
  month
  """, projectid, dialect='standard')
```

```python
#obtain data for highest 10 CCG prescribers of trimethoprim
highest_10_percent_df = pd.read_gbq("""
SELECT
  month,
  IEEE_DIVIDE(SUM(
      CASE
        WHEN SUBSTR(highest.bnf_code,1,9)='0501080W0' THEN items
        ELSE 0 END), SUM(items)) AS highest_10_trimeth_percent
FROM
  `ebmdatalab.richard.trimethoprim_data_highest_10` AS highest
WHERE
  month>='2014-01-01'
GROUP BY
  month
ORDER BY
  month
  """, projectid, dialect='standard')
```

```python
#plot data for both, and include vertical lines for key dates
plt.figure(figsize=(10,8))
plt.plot('month','lowest_10_trimeth_percent', data=lowest_10_percent_df)
plt.plot('month','highest_10_trimeth_percent' , data=highest_10_percent_df)
plt.axvline(x='2014-10-01',label='PHE guidance changed', linestyle=':', color='C3')
plt.axvline(x='2016-10-01',label='QP document first released', linestyle=':', color='C4')
plt.axvline(x='2017-04-01',label='QP monitoring period starts', linestyle=':', color='C5')
plt.ylim(ymin=0)
plt.xlim(xmax='2017-12-01')
plt.legend(bbox_to_anchor=(0., -.3, 1., .102), loc=8,
       ncol=2, mode="expand", borderaxespad=0.)
```

```python
#obtain baseline percentage data for lowest 10 CCG prescribers of trimethoprim
ccg_lowest_percent_df = pd.read_gbq("""
SELECT
  pct,
  SUM(CASE
      WHEN SUBSTR(trimethoprim.bnf_code,1,9)='0501080W0' AND trimethoprim.month BETWEEN '2014-04-01 00:00:00' AND '2014-09-01 00:00:00' THEN items
      ELSE 0 END) AS base_trimeth_items,
  SUM(CASE
      WHEN trimethoprim.month BETWEEN '2014-04-01 00:00:00' AND '2014-09-01 00:00:00' THEN items
      ELSE 0 END) AS base_all_items,
  IEEE_DIVIDE(SUM(
      CASE
        WHEN SUBSTR(trimethoprim.bnf_code,1,9)='0501080W0' AND trimethoprim.month BETWEEN '2014-04-01 00:00:00' AND '2014-09-01 00:00:00' THEN items
        ELSE 0 END), SUM(CASE
        WHEN trimethoprim.month BETWEEN '2014-04-01 00:00:00' AND '2014-09-01 00:00:00' THEN items
        ELSE 0 END)) AS base_trimeth_percent,
  SUM(CASE
      WHEN SUBSTR(trimethoprim.bnf_code,1,9)='0501080W0' AND trimethoprim.month BETWEEN '2014-10-01 00:00:00' AND '2015-03-01 00:00:00' THEN items
      ELSE 0 END) AS PHC_trimeth_items,
  SUM(CASE
      WHEN trimethoprim.month BETWEEN '2014-10-01 00:00:00' AND '2015-03-01 00:00:00' THEN items
      ELSE 0 END) AS PHE_all_items,
  IEEE_DIVIDE(SUM(
      CASE
        WHEN SUBSTR(trimethoprim.bnf_code,1,9)='0501080W0' AND trimethoprim.month BETWEEN '2014-10-01 00:00:00' AND '2015-03-01 00:00:00' THEN items
        ELSE 0 END), SUM(CASE
        WHEN trimethoprim.month BETWEEN '2014-10-01 00:00:00' AND '2015-03-01 00:00:00' THEN items
        ELSE 0 END)) AS PHE_trimeth_percent,
  SUM(CASE
      WHEN SUBSTR(trimethoprim.bnf_code,1,9)='0501080W0' AND trimethoprim.month BETWEEN '2017-04-01 00:00:00' AND '2017-09-01 00:00:00' THEN items
      ELSE 0 END) AS QP_trimeth_items,
  SUM(CASE
      WHEN trimethoprim.month BETWEEN '2017-04-01 00:00:00' AND '2017-09-01 00:00:00' THEN items
      ELSE 0 END) AS QP_all_items,
  IEEE_DIVIDE(SUM(
      CASE
        WHEN SUBSTR(trimethoprim.bnf_code,1,9)='0501080W0' AND trimethoprim.month BETWEEN '2017-04-01 00:00:00' AND '2017-09-01 00:00:00' THEN items
        ELSE 0 END), SUM(CASE
        WHEN trimethoprim.month BETWEEN '2017-04-01 00:00:00' AND '2017-09-01 00:00:00' THEN items
        ELSE 0 END)) AS QP_trimeth_percent
FROM
  `ebmdatalab.richard.trimethoprim_data_lowest_10` AS trimethoprim
INNER JOIN
  ebmdatalab.hscic.ccgs AS ccg
ON
  trimethoprim.pct=ccg.code
GROUP BY
  pct
ORDER BY
  pct
""", projectid, dialect='standard')
```

```python
ccg_lowest_percent_df.aggregate(['sum', 'min','max'])
```
